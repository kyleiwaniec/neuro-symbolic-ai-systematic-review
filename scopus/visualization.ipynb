```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
font = {'family' : 'times',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
dataset = pd.read_csv("Final_41 - Final-papers.csv")
```

```python
fig, ax = plt.subplots()
kautzCount = pd.value_counts(dataset['Kautz category'].values, sort=True)
ind = np.arange(len(kautzCount.index))
plt.style.use('seaborn')
ax.barh(kautzCount.index, kautzCount.values, color='indianred',  align='center')

fig.suptitle('Kautz category', fontsize=18)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Category', fontsize=14)
ax.set_yticklabels(kautzCount.index, minor=False)
for i, v in enumerate(kautzCount.values):
    ax.text(v+0.3, i + .1, str(v), fontdict=dict(fontsize=14))
    
plt.savefig('Kautz category.jpg', bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
docType = pd.value_counts(dataset['Doc type'].values, sort=True)
width = 0.8       # the width of the bars

rect=docType.plot.bar(color='peru', width=width)
ind = np.arange(len(docType.index))

fig.suptitle('Document type', fontsize=18)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Document', fontsize=14)

ax.set_xticklabels(docType.index, minor=False, fontsize=16, rotation=0)

for index,data in enumerate(docType.values):
    ax.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=14))

plt.savefig('docType.jpg',  bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
year = pd.value_counts(dataset['Year'].values, sort=True)

width = 0.8       # the width of the bars

rect=year.plot.bar(color='teal', width=width)
ind = np.arange(len(year.index))

fig.suptitle('Year of publication distribution', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Count', fontsize=14)

ax.set_xticklabels(year.index, minor=False, fontsize=16, rotation=0)

for index,data in enumerate(year.values):
    ax.text(x=index , y =data-0.8, s=f"{data}" , fontdict=dict(fontsize=14))

plt.savefig('year.jpg',  bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
tol = pd.value_counts(dataset['Type of Learning'].values, sort=True)

width = 0.6       # the width of the bars

rect=tol.plot.bar(color='royalblue', width=width)
ind = np.arange(len(tol.index))

fig.suptitle('Type of learning', fontsize=18)
plt.xlabel('Learning', fontsize=14)
plt.ylabel('Count', fontsize=14)

ax.set_xticklabels(tol.index, minor=False, fontsize=16, rotation=20)

for index,data in enumerate(tol.values):
    ax.text(x=index , y =data-1.2, s=f"{data}" , fontdict=dict(fontsize=14))

plt.savefig('tolearning.jpg',  bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
cta = pd.value_counts(dataset['Technical Application'].values, sort=True)
ind = np.arange(len(cta.index))
plt.style.use('seaborn')
ax.barh(cta.index, cta.values, color='crimson',  align='center')

fig.suptitle('Count of Technical Application', fontsize=24)
plt.xlabel('Count', fontsize=18)
plt.ylabel('Techincal application', fontsize=18)
ax.set_yticklabels(cta.index, minor=False)
for i, v in enumerate(cta.values):
    ax.text(v+0.3, i + .1, str(v), fontdict=dict(fontsize=14))
    
plt.savefig('ctechapplication.jpg',  bbox_inches='tight')
```
