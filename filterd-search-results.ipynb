```python
import numpy as np
import pandas as pd
```

```python
listdf=pd.read_csv("CIteScore-computer-science-journals.csv")
```

```python
listdf.columns
```

```python
narrowListDF=listdf[['Scopus Source ID', 'Title', 'Publisher', 'Scopus Sub-Subject Area', 'Open Access', 'Citation Count', 
       'CiteScore 2020', 'SNIP', 'SJR']]
```

```python
sorted(narrowListDF['Scopus Sub-Subject Area'].unique())
```

## Top 20 Publishers by CiteScore

```python
narrowListDF.loc[narrowListDF.groupby('Publisher')['CiteScore 2020'].idxmax(axis=1)].sort_values(by=['CiteScore 2020'], ascending=False).head(20)
```

## Top 20 Publishers by SJR

```python
narrowListDF.loc[narrowListDF.groupby('Publisher')['SJR'].idxmax()].sort_values(by=['SJR'], ascending=False).head(20)
```
